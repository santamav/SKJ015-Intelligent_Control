# Testing the DNN model

### Imports

```python
from machinevisiontoolbox.base import *
from machinevisiontoolbox import *
from spatialmath.base import *
from spatialmath import *

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R
```

### Prepare variables

```python
from VSM_PBVS import VSM_PBVS as V_PBVS

camera = CentralCamera.Default(pose=SE3.Trans(1, 1, -2))
P = mkgrid(2, 0.5)
p = camera.project_point(P, objpose=SE3.Tz(1))
T_Cd_G = SE3.Tz(1)

pbvs = V_PBVS(camera, P=P, pose_g=SE3.Trans(-1, -1, 2), pose_d=T_Cd_G, plotvol=[-2, 2, -2, 2, 0, 4])
```

```python
pbvs.run(100)
```

```python
std_pbvs = PBVS(camera, P=P, pose_g=SE3.Trans(-1,-1,2), pose_d=T_Cd_G, plotvol=[-2, 2, -2, 2, 0, 4])
std_pbvs.run(100)
```

### Compare results

#### Trajectory

```python
# Neural Network Controller PBVS
pbvs.plot_p()   
```

```python
# Original PBVS
std_pbvs.plot_p()
```

#### Pose

```python
# Neural Network Controller PBVS
pbvs.plot_pose()
```

```python
# Original PBVS
std_pbvs.plot_pose()
```

#### Velocity

```python
# Neural Network Controller PBVS
pbvs.plot_vel()
```

```python
# Original PBVS
std_pbvs.plot_vel()
```

### Evaluate errors

```python
# Extract the list 'p' from the history
p_pbvs = [hist.vel for hist in pbvs.history]
p_std_pbvs = [hist.vel for hist in std_pbvs.history]

# Compute rms between the two methods
rms = np.sqrt(np.mean((np.array(p_pbvs) - np.array(p_std_pbvs))**2))
print(rms)
```
